```python
# Importer les bibliothèques nécessaires
import pandas as pd
import os

# Afficher le répertoire de travail actuel pour vérifier que les chemins sont corrects
print("Répertoire de travail actuel :", os.getcwd())

# Définir les chemins des fichiers de données
movie_matrix_path = "/mnt/c/Users/edouard/Documents/GitHub/fev24_cmlops_movie_reco/data/processed/movie_matrix.csv"
user_matrix_path = "/mnt/c/Users/edouard/Documents/GitHub/fev24_cmlops_movie_reco/data/processed/user_matrix.csv"

# Initialiser les variables pour éviter NameError en cas d'échec de chargement
movie_matrix, user_matrix = None, None

# Charger les données
try:
    movie_matrix = pd.read_csv(movie_matrix_path)
    user_matrix = pd.read_csv(user_matrix_path)
    print("Fichiers chargés avec succès. Voici les premières lignes :")
    print("Movie Matrix:")
    print(movie_matrix.head())
    print("\nUser Matrix:")
    print(user_matrix.head())
except FileNotFoundError as e:
    print("Erreur de fichier non trouvé:", e)
except Exception as e:
    print("Une erreur s'est produite lors du chargement des fichiers:", e)

# S'assurer que les DataFrames sont chargés avant de continuer
if movie_matrix is not None and user_matrix is not None:
    # Explorer plus en détail les données
    # Afficher les statistiques descriptives pour 'movie_matrix'
    print("\nStatistiques descriptives de Movie Matrix:")
    print(movie_matrix.describe())

    # Afficher les informations sur les types de données et les valeurs manquantes pour 'user_matrix'
    print("\nInformations sur User Matrix:")
    print(user_matrix.info())

    # Exemple de jointure des deux matrices si elles ont une colonne en commun, par exemple 'user_id'
    if 'user_id' in movie_matrix.columns and 'user_id' in user_matrix.columns:
        combined_data = pd.merge(movie_matrix, user_matrix, on='user_id', how='inner')
        print("\nDonnées combinées après jointure sur 'user_id':")
        print(combined_data.head())
else:
    print("\nImpossible de charger les matrices de données pour l'analyse.")
```
